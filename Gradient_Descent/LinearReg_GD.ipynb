```python
import numpy as np
from sklearn.model_selection import cross_val_score
r2 = np.mean(cross_val_score(reg,X,y,scoring="r2",cv = 10))
r2
```

### Doing it for both m (slope) and b(intercept)

```python
# doing it for 2D line only for now
import numpy as np

class GD_Regressor:
    def __init__(self, epochs, lr):
        self.m = 100             # chose a random number this time for both m and b
        self.b = -120
        self.lr = lr
        self.epochs = epochs

    def fit(self,X,y):

        for i in range(self.epochs):
            loss_slope_b = -2*np.sum(y - self.m*X.ravel() - self.b)
            loss_slope_m = -2*np.sum((y - self.m*X.ravel() - self.b)*X.ravel())

            step_b = loss_slope_b*self.lr
            step_m = loss_slope_m*self.lr

            self.b = self.b - step_b
            self.m = self.m - step_m
            
            # print(self.b,self.m)

    def predict(self,X):
        return self.m * X + self.b

gd_reg = GD_Regressor(100,0.001)
# problem here is that choosing step value is very much trial and error

gd_reg.fit(X,y)
# gd_reg.predict(X)
gd_reg.m,gd_reg.b

# we are getting very close results to the sklearn result

```

```python
# as here cross_val_score will not work because that requires "get_params" method to be in Regression class which we have not coded in GD_Regressor
# so we are using train test split instead

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)

reg = GD_Regressor(100,0.001)
reg.fit(X_train,y_train)
# print(reg.m,reg.b)

y_pred = reg.predict(X_test)

from sklearn.metrics import r2_score
r2_score(y_test,y_pred)

# this also matched the linear regression part of r2 score
```
