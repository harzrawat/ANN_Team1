## Implementation of code from scratch for Logistic Regression using Perceptron-trick

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)


def perceptron_trick(X,y):
    wt = np.ones(len(X[0])+1)

    learning_rate = 0.1

    for i in range(1000):
        
        idx = random.choice(range(len(X)))      # bcz of random, the final line will most likely be different each time
        x= X[idx]
        y_real = y[idx]
        
        pt_matrix = np.insert(x,0,1)

        y_pred = np.dot(wt, pt_matrix)     # if f(x) < 0 then it will give 0 and if >0 then it will give 1

        if y_pred>0 and y_real==0:
            wt = wt - learning_rate*pt_matrix
        elif y_pred<0 and y_real==1:
            wt = wt + learning_rate*pt_matrix
        else:
            continue
        
        # this if else condition is combined in same equation in below code implementation
        
    return wt

perceptron_trick(X,y)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y,s=100)

c,a,b = perceptron_trick(X,y)

y_vals = np.linspace(-3,3,100)        # this can only be determined after plotting graph roughly at first as if the line is vertical then y range has to be defined as it can be very large even if we define x range 1st and then find y range
x_vals = -(b * y_vals + c)/a

plt.plot( x_vals, y_vals, color='red')

        
```

```python
# we can use sympy or we can define our own function to solve any expression
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)

def step(z):
    return 1 if z>0 else 0

def perceptron_trick(X,y):
    wt = np.ones(len(X[0])+1)

    learning_rate = 0.1

    for i in range(1000):
        
        idx = random.choice(range(len(X)))      # bcz of random, the final line will most likely be different each time
        x= X[idx]
        y_real = y[idx]

                
        # pt_matrix = np.concatenate(( [1], x.flatten() ))
        
        pt_matrix = np.insert(x,0,1)

        y_pred = step(np.dot(wt, pt_matrix))     # if f(x) < 0 then it will give 0 and if >0 then it will give 1

        wt = wt + learning_rate*pt_matrix*(y_real-y_pred)
        
    return wt

perceptron_trick(X,y)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y,s=100)

c,a,b = perceptron_trick(X,y)

y_vals = np.linspace(-3,3,100)        # this can only be determined after plotting graph roughly at first as if the line is vertical then y range has to be defined as it can be very large even if we define x range 1st and then find y range
x_vals = -(b * y_vals + c)/a

plt.plot( x_vals, y_vals, color='red')




        

        
```

## To improve our iplementation closer to sklearn
### Here we will be using sigmoid function, which will be used to push line away even from a point on correct side but not that far

```python
# making some changes here to improve the classification
# instead of step function we will be directly dealing with substituted results

import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)

import math
def sigmoid(z):
    return 1/(1+math.exp(-z))

def perceptron_trick_sigmoid(X,y):
    wt = np.ones(len(X[0])+1)

    learning_rate = 0.1

    for i in range(1000):
        
        idx = random.choice(range(len(X))) 
        x= X[idx]
        y_real = y[idx]
        
        pt_matrix = np.insert(x,0,1)

        y_pred = sigmoid(np.dot(wt, pt_matrix))

        wt = wt + learning_rate*pt_matrix*(y_real-y_pred)
        
    return wt

perceptron_trick_sigmoid(X,y)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y,s=100)

c,a,b = perceptron_trick_sigmoid(X,y)

y_vals = np.linspace(-3,3,100)        # this can only be determined after plotting graph roughly at first as if the line is vertical then y range has to be defined as it can be very large even if we define x range 1st and then find y range
x_vals = -(b * y_vals + c)/a

plt.plot( x_vals, y_vals, color='red')





        

        
```

### The above improvement is yet far behind of scikit learn result, so need to improve further

#### so to tackle this problem, we come up with concept of loss function:- but why
#### -- because, by finding loss function we can go on further trying to get optimised solution (line in this case), however there is no closed form solution( means any exact answer) so we will be using gradient descent (which is an approximate optimised solution ) to find the resultant more optimised line

```python
# making some changes here to improve the classification
# instead of step function we will be directly dealing with substituted results

import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)

import math
def sigmoid(z):
    return 1/(1+math.exp(-z))

def lossFunction(wt):
    res=0
    for i in range(len(X)):
        x_y_1 = np.insert(X[i],0,1)

        y_pred = sigmoid(np.dot(wt, x_y_1))
        res+=( (1-y[i]) * math.log( 1-sigmoid (y_pred) ) )

    return res

def perceptron_trick_sigmoid(X,y):
    wt = np.ones(len(X[0])+1)

    learning_rate = 0.1

    
    for i in range(1000):
        
        idx = random.choice(range(len(X))) 
        x= X[idx]
        y_real = y[idx]
        
        pt_matrix = np.insert(x,0,1)

        y_pred = sigmoid(np.dot(wt, pt_matrix))
        wt2 = wt + learning_rate*pt_matrix*(y_real-y_pred)
        if lossFunction(wt)>=lossFunction(wt2):
            wt = wt2
        else:
            break        
        
    return wt

perceptron_trick_sigmoid(X,y)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y,s=100)

c,a,b = perceptron_trick_sigmoid(X,y)

y_vals = np.linspace(-3,3,100)        # this can only be determined after plotting graph roughly at first as if the line is vertical then y range has to be defined as it can be very large even if we define x range 1st and then find y range
x_vals = -(b * y_vals + c)/a

plt.plot( x_vals, y_vals, color='red')





        

        
```
